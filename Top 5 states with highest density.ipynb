## Find top five states of US having highest population density in year of 2010.

```python
import numpy as pd
import pandas as pd
```

```python
url1="https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-population.csv"
url2="https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-areas.csv"
url3="https://raw.githubusercontent.com/jakevdp/data-USstates/master/state-abbrevs.csv"
```

```python
pop = pd.read_csv(url1)
areas = pd.read_csv(url2)
abbrevs = pd.read_csv(url3)
```

```python
pop.head()
```

```python
 areas.head()
```

```python
abbrevs.head()
```

```python
m1=pop.merge(abbrevs,left_on='state/region',right_on='abbreviation')
m1
```

```python
m1.drop('abbreviation',1)
```

```python
m1.isnull().any()
```

```python
final = pd.merge(m1, areas, on='state', how='left')
final.head()
```

```python
final.isnull().any()
```

```python
final[final['year']==2010].head()
```

```python
final['density']=final['population']/final['area (sq. mi)']
```

```python
final
```

```python
final[(final['year']==2010)&(final['ages']=='total')][['state','density']].sort_values(ascending=False,by='density').head()
```
